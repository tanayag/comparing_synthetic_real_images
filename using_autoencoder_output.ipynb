```python
import torch
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
ae_real_output = torch.load("real_output_encoder.pkl")
ae_synthetic_output = torch.load("synthetic_output_encoder.pkl")
```

```python
def convert_tensor(data):
    data = [i.detach().numpy() for i in data]
    data = np.concatenate(data, axis=0)
    final=[]
    for i in data:
        final.append(i.ravel())
    return final
```

```python
ae_real_output = convert_tensor(ae_real_output)
```

```python
ae_synthetic_output = convert_tensor(ae_synthetic_output)
```

```python
pca = PCA(n_components=2)
```

```python
reduced_real = pca.fit_transform(ae_real_output)
```

```python
reduced_synthetic = pca.fit_transform(ae_synthetic_output)
```

```python
def plot_scatter(data):
    x = data[:,0]
    y = data[:,1]
    axes = plt.gca()
    axes.set_xlim([-0.2,0.3])
    axes.set_ylim([-0.2,0.3])
    plt.scatter(x,y)
    plt.show()
```

```python
plot_scatter(reduced_real)
```

```python
plot_scatter(reduced_synthetic)
```
